```python
import pandas as pd
import glob
```

```python
files=glob.glob("./Files/*.txt") # match all txt files in the direcctory and return as a list
```

```python
files
```

```python
#files=['2021_11_30.txt','2021_12_04.txt', '2021_12_05.txt'] #list for file names
with open ('merged_file.txt', 'w') as new_file: # Open file3 in write mode
    
    for f_name in files: # Iterate through list
        with open(f_name) as file:           # Open each file in read mode
            new_file.write(file.read())       # read the data from files and save in mergedfiles
        new_file.write('\n') # nextline 
```

```python
with open('merged_file.txt') as f:
    contents = f.read()
    print(contents)
```

```python
with open('merged_file.txt') as f:
    lines= f.readlines() # reads each line in the file line by line
    with open('newfile.txt', 'w') as fw: #A new file
            for line in lines:
                if line.strip('\n') !='----------NEW TRANSACTON-----------': #reads all lines except --New transaction
                    fw.write(line)
```

```python
with open('newfile.txt') as f:
    contents = f.read()
    print(contents)
```

```python
b={} #empty dictionary
with open("newfile.txt","r") as f:
    for i in f.readlines(): #Read each line in the file
        if i.isspace():
            continue; #Skip line if blank
        a=i.strip().split(':', 1) #strip for removing whitespaces,Split data by first occurence of ':'
        print(a) 
        #Add data to dictionary simple hash algorithm
        if a[0] not in b.keys():
             b[a[0]] = [a[1]] #reads a[0] and the value in index 1
        else:
            b[a[0]].append(a[1]) # append index 1
        
```

```python
b
```

```python
df=pd.DataFrame(b)
df.head()
```

```python
df.columns = df.columns.str.title() #set columns to proper case
```

```python
df.columns=df.columns.str.replace (' ', '') # Remove whitespace
```

```python
#Remove whitespace from all columns in the data set
for column in df.columns:
    df[column]=df[column].str.strip()
    #df[column].apply(lambda x: x.str.strip())
```
